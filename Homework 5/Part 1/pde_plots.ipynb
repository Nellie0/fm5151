```python
import numpy as np
import matplotlib.pyplot as plt
from pde import BSM_PDE
```

```python
S = 100
K = 102
r = 0.05
q = 0.01
T = 1
sigma = 0.2
M = 100
N = 1000
BSM = BSM_PDE(S, K, r, q, sigma, T, M, N)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

S_mesh, T_mesh = np.meshgrid(BSM[0], BSM[1])

ax.plot_surface(S_mesh, T_mesh, BSM[2])
ax.view_init(30, -100) #rotate graph
```
